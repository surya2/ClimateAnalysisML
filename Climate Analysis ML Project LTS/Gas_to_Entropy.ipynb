```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
finalCO2 = './Datasets/Programmed Datasets/FinalCO2.csv'
finalCO2 = pd.read_csv(finalCO2)
CO2 = finalCO2.interpolatedCO2
CO2 = CO2[70:]

CO2year = finalCO2.Year
CO2time = finalCO2.time
features = ['Year', 'time']
CO2x = finalCO2[features]
```

```python
finalSST = './Datasets/Programmed Datasets/FinalSST.csv'
finalSST = pd.read_csv(finalSST)

SSTyear = finalSST.Year
SSTtime = finalSST.time
SST = finalSST.extrapSST
print(len(SST))
```

```python
import math
z = 18000
e = math.e
pstrat = 22.65*(e)**(1.73 - (0.000157 * z))
r = 8.314
```

```python
# according to Clark. W. C., 1 ppm C = 2.31 Gt C
import math

mass_of_CO2 = list()
for i in CO2:
    mass_gt = i * 2.31
    mass_grams = mass_gt * (math.pow(10, 15))
    mass_of_CO2.append("{:.2e}".format(mass_grams))

mol_of_CO2 = list()
for o in mass_of_CO2:
    mol_of_CO2.append(float(o)/44.01)
```

```python
entropy_of_CO2 = list()
for s in SST:
    entropy_of_CO2.append((28.2 * math.log(((float(s) + 273) - (0.0065 * 11000)) / (float(s) + 273))) + (r * math.log(101.29/pstrat)))

#print(len(co2entropy))

holistic_entropy_CO2 = list()
for x in range(len(mol_of_CO2)):
    holistic_entropy_CO2.append(mol_of_CO2[x] * entropy_of_CO2[x])
    
print(holistic_entropy_CO2)
```

```python
data = {'Year': CO2year,
        'time': CO2time,
        'entropy': holistic_entropy_CO2
        }
holistic_entropy = pd.DataFrame({"Year":SSTyear, 
                    "time":SSTtime,  
                    "entropy":holistic_entropy_CO2}) 
holistic_entropy.to_csv('./Datasets/Programmed Datasets/Entropy.csv')
```
