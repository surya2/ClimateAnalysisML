```python
from scipy.interpolate import interp1d

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
#import xgboost as xgb
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import kendalltau
from scipy.stats import spearmanr
from scipy.stats import pointbiserialr
import math
```

```python
data = 'Climate Change Analysis/Datasets/Stage 2/Air-Sea_Flux_to_Atlantic_Cyclone.csv'
cycloneData = pd.read_csv(data)
#print(cycloneData)

sst = cycloneData.SST
```

```python
x = cycloneData.Year
time = cycloneData.time
features = ['Year', 'time']
X = cycloneData[features]
```

```python
f = interp1d(time, sst)
f2 = interp1d(time, sst, kind='cubic')
print(time[69])
xnew = np.linspace(0, time[69], num=41, endpoint=True)
print(xnew)
plt.plot(time, sst, 'o', time, f(time), '-', time, f2(time), '--')
plt.legend(['data', 'linear', 'cubic'], loc='best')
plt.show()
```
